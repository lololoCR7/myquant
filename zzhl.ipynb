```python
index_merge=pd.merge(index, bench, on='日期')
```

```python
x='sum_20_x'
y='sum_20_y'
```

```python
index_updown=index_merge[['日期',x,y]]
```

```python
index_updown=index_updown.dropna(subset=[x,y])
index_updown
```

```python
startdate='2018-01-01'
```

```python
A = index_updown[index_updown['日期']>=startdate][x].tolist()
B = index_updown[index_updown['日期']>=startdate][y].tolist()
```

```python
# stats.normaltest
statistic, p_value = scipy.stats.normaltest (A, axis=0)
print('A - 统计量：', statistic)
print('A - P值：', p_value)
if p_value>0.05:
    print('A服从正态分布')
else:
    print('A不服从正态分布')

statistic, p_value = scipy.stats.normaltest (B, axis=0)
print('B - 统计量：', statistic)
print('B - P值：', p_value)
if p_value>0.05:
    print('B服从正态分布')
else:
    print('B不服从正态分布')
```

```python
z,p=sw.ztest(A,B,value=0,alternative="larger")#{“two-sided”, larger, smaller}
# z,p=sw.ztest(A,value=0,alternative="smaller")#{“two-sided”, larger, smaller}
print("z值={},p值={}".format(z,p))
```
